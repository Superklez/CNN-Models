```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from skimage import io
from torchvision.transforms import transforms
```

```python
filenames = []
labels = []

for filename in os.listdir(os.getcwd() + '/train'):
    
    if filename[:3] == 'cat':
        filenames.append(filename)
        labels.append(0)
        
    elif filename[:3] == 'dog':
        filenames.append(filename)
        labels.append(1)
        
df = pd.DataFrame(filenames, columns=['filename'])
df['label'] = labels
df.head(10)
```

```python
num_cats = 250
num_dogs = 250

df_cats = df[df['label'] == 0]
df_dogs = df[df['label'] == 1]

df_cats_sampled = df_cats.sample(num_cats)
df_dogs_sampled = df_dogs.sample(num_dogs)

df_sampled = df_cats_sampled.append(df_dogs_sampled)
df_sampled = df_sampled.sample(frac=1).reset_index(drop=True)

df_sampled.head(10)
```

```python
df_sampled.to_csv('train.csv', index=False)
```
